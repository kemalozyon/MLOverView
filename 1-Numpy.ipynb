**BroadCasting in NumPy:**

```python
arr[0:5] = 100
```

```python
arr
```

```python
arr = np.arange(0,11)
```

```python
slice_of_arr = arr[0:5] # This will point to the same address anything we do in slice_of_arr also change our original array: call by reference
```

```python
arr
```

```python
slice_of_arr
```

```python
slice_of_arr[:] = 99
```

```python
slice_of_arr
```

```python
arr
```

```python
# To prevent this
slice_of_arr2 = arr[0:5].copy()
```

```python
slice_of_arr2[:] = 11
```

```python
slice_of_arr2
```

```python
arr # arr doesnt change because slice_of_arr2 is call by value variable
```
