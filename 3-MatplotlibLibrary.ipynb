```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Two main goals with matplotlib**
- Be able to plot out a functional relationship
    - y = f(x) = x^2
- Be able to plot out data points
    - (x1, y1), (x2, y2), (x3, y3), ...

```python
a = np.linspace(0, 10, 11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[1].plot(a, b)
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3)

axes[0][0].plot(x, y)
axes[1, 1].plot(a,b)
axes[2,2].plot(a,b)
axes[1,1].set_xlabel("I am in the middle")
axes[1,1].set_ylabel("I am in the middle")
fig.subplots_adjust(wspace=0.9, hspace=0.9)
fig.suptitle("Figure Level", fontsize=20)
fig.set_figwidth(10)
fig.set_figheight(5)

plt.show()
```

```python
fig = plt.figure(figsize=(4,3))
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([1.5,0,1,1])
axes1.plot(x, x, label="x vs x")
axes1.plot(x, x**2, label="x vs x^2")
axes2.plot(x, x)
axes2.set_ylim(0,10)
axes1.legend(loc=0)
axes2.set_xlabel("Zoom in")
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x, color="orange", label="x vs x", lw=3, ls=":")
ax.plot(x, x + 1, color="purple", label="x vs x + 1",lw=2, marker="o")
ax.set_ylim(0, 12)
ax.set_xlim(0, 9)
ax.legend(loc=2)

plt.show()
```
